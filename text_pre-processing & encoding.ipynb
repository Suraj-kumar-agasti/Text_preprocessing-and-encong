```python
# remove html tags
import re

def remove_html_tags(series):
    pattern = re.compile('<.*?>')
    return series.apply(lambda x: pattern.sub("", str(x)))

df["text"] = remove_html_tags(df["text"])
```

```python
remove_html_tags(df["text"])
```

```python
# remove url
import re

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("", text)

df["text"] = df["text"].apply(remove_url)
```

```python
df.head()
```

```python
# remove punctuations
import string
exclude = string.punctuation
chat_words={
   " AFAIK":"As Far As I Know",
"AFK": "Away From Keyboard",
"ASAP":"As Soon As Possible",
"BTW":"By The Way",
"B4":"Before",
"LAMO":"Laugh My A.. Off",
"FYI":"For your information"
}
```

```python
def chat_conversion(text):
    new_text=[]
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
def remove_punc(text):
    for char in exclude:
        text = text.replace(char,"")
    return text
```

```python
df["text"] = df["text"].apply(remove_punc)
```

```python
def remove_emojis_manually(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F700-\U0001F77F"  # alchemical symbols
                               u"\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
                               u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
                               u"\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
                               u"\U0001FA00-\U0001FA6F"  # Chess Symbols
                               u"\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
                               u"\U00002702-\U000027B0"  # Dingbats
                               u"\U000024C2-\U0001F251"
                               "]+", flags=re.UNICODE)
    clean_text = emoji_pattern.sub(r'', text)
    return clean_text
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
def remove_stopwords(text):
    new_text=[]


    for word in text.split():
        if word in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())




    return " ".join(new_text).replace("   ","")
```

```python
"""Stemming vs Lemmatization
Stemming is a process that stems or removes last few characters from a word, often leading to incorrect meanings and spelling.
For instance, stemming the word â€˜Caringâ€˜ would return â€˜Carâ€˜.
Stemming is used in case of large dataset where performance is an issue.

Lemmatization considers the context and converts the word to its meaningful base form, which is called Lemma.
For instance, lemmatizing the word â€˜Caringâ€˜ would return â€˜Careâ€˜.
Lemmatization is computationally expensive since it involves look-up tables and what not."""
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word
```

```python
""""
The IDF (Inverse Document Frequency) and TF-IDF (Term Frequency-Inverse Document Frequency) are techniques commonly used in information retrieval and text mining tasks. Here are the advantages and disadvantages of each:

Advantages of IDF:

Reflects the importance of rare terms: IDF assigns higher weights to terms that are rare across the corpus but appear in specific documents. This helps in distinguishing between common terms and those that are more informative.

Simple to compute: IDF is relatively simple to compute and understand. It only requires calculating the logarithm of the ratio of total documents to the number of documents containing a specific term.

Disadvantages of IDF:

Sensitivity to corpus size: IDF can be sensitive to the size and composition of the corpus. In very large or very small corpora, the IDF values might not reflect the true importance of terms accurately.

No consideration of term frequency: IDF alone does not consider the frequency of terms within documents, which can be crucial for understanding their significance in a particular context.

Advantages of TF-IDF:

Combines term frequency and inverse document frequency: TF-IDF integrates both the frequency of a term within a document (TF) and its rarity across the corpus (IDF), providing a more nuanced measure of term importance.

Effective in handling stop words: TF-IDF naturally discounts the influence of common words (stop words) by giving them lower weights due to their high frequency in most documents.

Flexible weighting scheme: TF-IDF allows for customization through different weighting schemes (e.g., sublinear TF scaling, smoothing techniques), making it adaptable to various text mining tasks.

Disadvantages of TF-IDF:

Vulnerable to document length: Longer documents may have higher TF-IDF scores simply due to the increased occurrence of terms, which may not always reflect their true importance.

Requires parameter tuning: Selecting appropriate parameters, such as normalization methods or the handling of rare terms, can be non-trivial and may require experimentation.

Doesn't capture semantic similarity: TF-IDF treats terms independently and does not consider the semantic relationships between them, which could limit its effectiveness in tasks requiring deeper understanding of language semantics.

Overall, while IDF and TF-IDF are powerful techniques for term weighting and information retrieval, they have their own strengths and weaknesses that should be considered based on the specific requirements of the NLP task at hand.




""""
```
